# Plot of Missing data

```r
# Show pattern of Missing data
```

```r
# A function that plots missingness
# requires `reshape2`

library(reshape2)
library(ggplot2)

ggplot_missing <- function(x){
  
  x %>% 
    is.na %>%
    melt %>%
    ggplot(data = .,
           aes(x = Var2,
               y = Var1)) +
    geom_raster(aes(fill = value)) +
    scale_fill_grey(name = "",
                    labels = c("Present","Missing")) +
    theme_minimal() + 
    theme(axis.text.x  = element_text(angle=45, vjust=0.5)) + 
    labs(x = "Variables in Dataset",
         y = "Rows / observations")
}
```

```r
ggplot_missing(data_PM10)
```

```r
data = read.csv("ItemListNoDummy.csv", header = TRUE)
```

```r
data[is.na(data)] <- 0
```

```r
head(data, n=5);
```

```r
install.packages("ggplot2") # The mice package =====>>>> Multivariate Imputation via Chained Equations)
library(ggplot2)
```

```r
ggplot(data, aes(x=PM25, y=PM10)) +
    geom_point(shape=1)      # Use hollow circles

ggplot(data, aes(x=CO, y=PM10)) +
    geom_point(shape=1) +    # Use hollow circles
    geom_smooth(method=lm)   # Add linear regression line 
                             #  (by default includes 95% confidence region)

ggplot(data, aes(x=NO2, y=PM10)) +
    geom_point(shape=1) +    # Use hollow circles
    geom_smooth(method=lm,   # Add linear regression line
                se=FALSE)    # Don't add shaded confidence region


ggplot(data, aes(x=O3, y=PM10)) +
    geom_point(shape=1) +    # Use hollow circles
    geom_smooth()            # Add a loess smoothed fit curve with confidence region
#> `geom_smooth()` using method = 'loess'
```
